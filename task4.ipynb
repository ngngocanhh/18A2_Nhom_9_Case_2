```python
import pandas as pd
info = pd.read_csv("stock_info.csv")
price = pd.read_csv("stock_price.csv")
trans = pd.read_csv("transactions.csv")
```

```python
df_merge_1 = pd.merge(
    info,
    price,
    on="ticker",
    how="left",
    indicator=True
)
df_merge_1["_merge"].value_counts()
missing_price = df_merge_1[df_merge_1["_merge"] == "left_only"][["ticker"]].drop_duplicates()
missing_price

```

```python
df_final = pd.merge(
    df_merge_1.drop(columns=["_merge"]),
    trans,
    on="ticker",
    how="left",
    indicator=True
)
df_final["_merge"].value_counts()
missing_trans = df_final[df_final["_merge"] == "left_only"][["ticker"]].drop_duplicates()
missing_trans
```

```python
import pandas as pd
info = pd.read_csv("stock_info.csv")
price = pd.read_csv("stock_price.csv")
trans = pd.read_csv("transactions.csv")

print(info.head())
print(price.head())
print(trans.head())
df_merge_1 = pd.merge(
    info,
    price,
    on="ticker",
    how="left",
    indicator=True
)
df_merge_1["_merge"].value_counts()
missing_price = df_merge_1[df_merge_1["_merge"] == "left_only"][["ticker"]].drop_duplicates()
missing_price
df_final = pd.merge(
    df_merge_1.drop(columns=["_merge"]),
    trans,
    on="ticker",
    how="left",
    indicator=True
)
df_final["_merge"].value_counts()
missing_trans = df_final[df_final["_merge"] == "left_only"][["ticker"]].drop_duplicates()
missing_trans
df_final["volume"] = df_final["volume"].fillna(0)
df_final["trade_type"] = df_final["trade_type"].fillna("UNKNOWN")

# Xóa cột kiểm tra
df_final = df_final.drop(columns="_merge")

# ================================
# 5. Xuất file dữ liệu hoàn chỉnh
# ================================
df_final.to_csv(
    "stock_full_merged.csv",
    index=False
)
```
