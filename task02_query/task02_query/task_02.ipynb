```python
import pandas as pd

df_info = pd.read_csv('../data_clean/stock_info_clean.csv')
df_price = pd.read_csv('../data_clean/stock_price_clean.csv')
df_trans = pd.read_csv('../data_clean/transactions_clean.csv')
```

# TRUY VẤN VÀ THỐNG KÊ MÔ TẢ


```python
#LỌC DỮ LIỆU THEO SÀN GIAO DỊCH

df_hose = df_info[df_info["exchange"] == "HOSE"]
df_upcom = df_info[df_info["exchange"] == "UPCOM"]
df_hnx = df_info[df_info["exchange"] == "HNX"]

print(df_hose.head())
print(df_upcom.head())
print(df_hnx.head())
```

```python
# THỐNG KÊ SỐ LƯỢNG MÃ THEO TỪNG NGÀNH

df_info.groupby("sector")["ticker"].count()
```

```python
# CÁC MÃ CÓ GIÁ ĐÓNG CỬA BỊ THIẾU

missing_price = df_price[df_price["close_price"].isna()]
print(missing_price)
```

```python
# THỐNG KÊ KHỐI LƯỢNG GIAO DỊCH TRUNG BÌNH

# Trung bình theo mã cổ phiếu

avg_volume = df_trans.groupby("ticker")["volume"].mean()
avg_volume

```
