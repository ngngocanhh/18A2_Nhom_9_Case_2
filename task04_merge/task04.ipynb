```python
import pandas as pd

print("=== TASK 04 – MERGE DỮ LIỆU ===\n")

df_info  = pd.read_csv("../data_clean/stock_info_clean.csv")
df_price = pd.read_csv("../data_clean/stock_price_clean.csv")
df_trans = pd.read_csv("../data_clean/transactions_clean.csv")

print(">>> Đã đọc xong dữ liệu\n")
print("Cột df_info :", df_info.columns)
print("Cột df_price:", df_price.columns)
print("Cột df_trans:", df_trans.columns, "\n")
```

```python
print("=== Chuẩn hóa cột ngày trong PRICE ===")

if "trade_date" in df_price.columns:
    df_price = df_price.rename(columns={"trade_date": "date"})
elif "price_date" in df_price.columns:
    df_price = df_price.rename(columns={"price_date": "date"})

print("Cột df_price sau chuẩn hóa:", df_price.columns, "\n")
```

```python
print("df_price columns:", df_price.columns)
print("df_merge columns:", df_merge.columns)
print("df_trans columns:", df_trans.columns)
```

```python
print("=== Ý 1: MERGE GIÁ + THÔNG TIN CỔ PHIẾU ===")

df_merge = df_price.merge(
    df_info,
    on="ticker",
    how="left"
)

print(">>> df_merge sau merge info:")
print(df_merge.head(), "\n")
```

```python
print("=== Chuẩn hóa cột ngày trong TRANSACTIONS ===")

if "trade_date" in df_trans.columns:
    df_trans = df_trans.rename(columns={"trade_date": "date"})
elif "transaction_date" in df_trans.columns:
    df_trans = df_trans.rename(columns={"transaction_date": "date"})
elif "trans_date" in df_trans.columns:
    df_trans = df_trans.rename(columns={"trans_date": "date"})

print("Cột df_trans sau chuẩn hóa:", df_trans.columns, "\n")
```

```python
print("=== Ý 2: MERGE THÊM DỮ LIỆU GIAO DỊCH ===")

df_merge = df_merge.merge(
    df_trans,
    on="ticker",
    how="left"
)

print(">>> Merge transactions thành công")
print(df_merge.head())
```

```python
print("=== KIỂM TRA DỮ LIỆU SAU MERGE ===")

print("Số dòng:", len(df_merge))
print("Số cột :", df_merge.shape[1])

print("\nSố giá trị thiếu theo cột:")
print(df_merge.isna().sum(), "\n")
```

```python
df_merge.to_csv(
    "../task04_merge/stock_full_merge.csv",
    index=False
)

print("=== HOÀN THÀNH TASK 04 – ĐÃ TẠO stock_full_merge.csv ===")
```
